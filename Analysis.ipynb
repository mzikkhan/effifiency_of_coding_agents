```python
closed_prs = pr_df.loc[pr_df['state'] == 'closed', ['id', 'state']]
```

```python
import matplotlib.pyplot as plt

# Counts
counts = {
    "Closed": len(closed_prs),
    "Open": len(not_closed_prs)
}

# Bar chart
plt.figure(figsize=(6,4))
plt.bar(counts.keys(), counts.values())
plt.title("Closed vs Open Pull Requests")
plt.xlabel("PR Status")
plt.ylabel("Count")
plt.show()
```

```python
print(pr_reviews_df.columns)  

print(closed_prs.columns)  
```

```python
closed_with_reviews = closed_prs.merge(
    pr_reviews_df,
    left_on="id",
    right_on="pr_id",
    how="left"
)

```

```python
closed_with_reviews
```

```python
iteration_counts = closed_with_reviews.groupby("pr_id").size().reset_index(name="iterations")
```

```python
iteration_counts
```

```python
closed_with_iterations = closed_prs.merge(
    iteration_counts,
    left_on="id",
    right_on="pr_id",
    how="left"
)

closed_with_iterations['iterations'] = closed_with_iterations['iterations'].fillna(0)
```

```python
closed_with_iterations
```

```python
col = closed_with_iterations['iterations']

stats = {
    "mean": col.mean(),
    "median": col.median(),
    "mode": col.mode().tolist(),
    "min": col.min(),
    "max": col.max()
}

stats
```
