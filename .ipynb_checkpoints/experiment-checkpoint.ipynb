## Iteration Cycles

```python
pr_iteration_cycles = (
    pr_timeline_df
    .groupby("pr_id")
    .size()
    .reset_index(name="iteration_cycles")
)

pr_iteration_cycles
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(pr_iteration_cycles["iteration_cycles"], bins=30)
plt.xlabel("Number of Iteration Cycles per PR")
plt.ylabel("Number of PRs")
plt.title("Distribution")
plt.show()
```

```python
pr_iteration_cycles["iteration_cycles"].describe()
```

## PR Lifecycle (Time)

```python
pr_df.head()
```

```python
pr_df[pr_df["id"] == 2823956710]
```

```python
# Are closed_at timestamps the same as merged_at timestamps?

sample_pr_df = pr_df.copy()

sample_pr_df["closed_at"] = pd.to_datetime(sample_pr_df["closed_at"])
sample_pr_df["merged_at"] = pd.to_datetime(sample_pr_df["merged_at"])
sample_pr_df["merged_closed"] = sample_pr_df["closed_at"] - sample_pr_df["merged_at"]

sample_pr_df.head()
```

```python

merged_v_closed = (
    sample_pr_df
    .groupby("merged_closed")
    .size()
    .reset_index(name="merged_closed_counts")
)

merged_v_closed
```

```python
# Conclusion: Insignificant delay between merged_at timestamps and closed_at timestamps
```

```python
# Calculate lifecycle of each PR

pr_df_2 = pr_df.copy()

pr_df_2["created_at"] = pd.to_datetime(pr_df_2["created_at"])
pr_df_2["closed_at"] = pd.to_datetime(pr_df_2["closed_at"])

pr_df_2["lifecycle"] = pr_df_2["closed_at"] - pr_df_2["created_at"]
pr_df_2["lifecycle_hours"] = pr_df_2["lifecycle"].dt.total_seconds() / 3600

pr_df_2.head()
```

```python
pr_lifecycle = (
    pr_df_2
    .groupby("lifecycle_hours")
    .size()
    .reset_index(name="lifecycle_counts")
)

pr_lifecycle
```

```python
pr_lifecycle.describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(pr_lifecycle["lifecycle_counts"], bins=30, log=True)
plt.xlabel("Lifecycle Hours")
plt.ylabel("Counts (log scale)")
plt.title("Distribution (Log Scale)")
plt.show()
```

## Merging of 3 dimensions - Closed + Iteration Count + Lifecycle

```python
filtered_pr = pr_df_2[pr_df_2["state"] == "closed"].copy()
filtered_pr = filtered_pr[["id", "state", "lifecycle_hours"]]
filtered_pr
```

# Experiment 2 : Segmenting Reviewer Type - BOT vs Human

```python
pr_reviews_df.head()
```

```python
reviews_trimmed = pr_reviews_df[["pr_id", "user_type"]].copy()

merged_with_reviews = pd.merge(
    merged_df,
    reviews_trimmed,
    left_on="id",
    right_on="pr_id",
    how="left"
)
```

```python
merged_with_reviews = merged_with_reviews.drop(columns=["pr_id"])

merged_with_reviews
```

```python
merged_with_reviews["user_type"].value_counts(dropna=False)
```

```python
# 23,968 PRs have no review user_type information

# 48,985 rows vs 31,284 rows
```

```python
merged_with_reviews["id"].value_counts().describe()
```

```python
merged_with_reviews = merged_with_reviews.drop_duplicates()
merged_with_reviews
```

```python
duplicate_ids = merged_with_reviews["id"].value_counts()
duplicate_ids[duplicate_ids > 1]
```

```python
merged_with_reviews[merged_with_reviews["id"] == 3176255193]
```

```python
merged_with_reviews[merged_with_reviews["id"] == 3133002658]
```

```python
# Conclusion: some PRs have both user and BOT reviewers. We can designate them as a third entity.
```

```python
# Drop rows with no reviewer information

clean_reviews = merged_with_reviews.dropna(subset=["user_type"])
clean_reviews
```

```python
clean_reviews["user_type"].unique()
```

```python
clean_reviews[clean_reviews["id"] == 3133002658]
```

```python
# Create Both user_type

type_map = (
    clean_reviews.groupby("id")["user_type"]
    .apply(lambda x: set(x.unique()))
)


def label_user_type(types):
    if types == {"User"}:
        return "User"
    elif types == {"Bot"}:
        return "Bot"
    else:
        return "Both"

pr_type_labels = type_map.apply(label_user_type).reset_index()
pr_type_labels.columns = ["id", "user_type_final"]

clean_reviews_labeled = clean_reviews.merge(
    pr_type_labels,
    on="id",
    how="left"
)

clean_reviews_labeled
```

```python
# Keep only one unique row per PR id

final_df = clean_reviews_labeled.groupby("id").agg({
    "lifecycle_hours": "first",   
    "iteration_cycles": "first", 
    "user_type_final": "first" 
}).reset_index()

final_df
```

```python
final_df[final_df["id"] == 3265640341]
```

```python
# Get an idea of user_type wise summary stats on effiency

final_df.groupby("user_type_final")[["lifecycle_hours", "iteration_cycles"]].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x="user_type_final", y="lifecycle_hours", data=final_df)
plt.yscale("log")
plt.ylabel("Lifecycle Hours")
plt.xlabel("Reviewer Type")
plt.title("AI PR Lifecycle by Reviewer Type")
plt.show()
```

```python
# Bot reviews give shortest lifecycle ?
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="user_type_final", y="iteration_cycles", data=final_df)
plt.ylabel("Iteration Cycles")
plt.xlabel("Reviewer Type")
plt.title("AI PR Iteration Cycles by Reviewer Type")
plt.show()
```

```python
# Bot reviews give shortest Iteration cycles?
```

```python
final_df["user_type_final"].value_counts()
```

#### Conclusion : Bot generated reviews lead to most efficient PRs - shortest lifecycle (in hours) and smallest number of iteration cycles. Combination of User and Bot reviews lead to least efficiency.

#### However, it must be noted that number of data points of Bot generated reviews ONLY is fewer compared to others.

# Experiment 3 : Tone of Review Message - Positive vs Negative vs Neutral 

```python
merged_df
```

```python
pr_reviews_df.tail()
```

```python
reviews_extracted = pr_reviews_df[["pr_id", "body"]].copy()
reviews_extracted
```

```python
reviews_extracted = reviews_extracted.dropna(subset=["body"])
reviews_extracted
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

# Function to classify sentiment
def classify_sentiment_vader(text):
    score = sia.polarity_scores(text)["compound"]
    if score > 0.05:
        return 1
    elif score < -0.05:
        return -1
    else:
        return 0

reviews_extracted["sentiment_label"] = reviews_extracted["body"].apply(classify_sentiment_vader)

reviews_extracted.head()
```

```python
reviews_extracted
```

```python
reviews_extracted["sentiment_label"].value_counts()
```

```python
final_df_2 = pd.merge(
    merged_df,
    reviews_extracted,
    left_on="id",
    right_on="pr_id",
    how="left"
)

final_df_2
```

```python
final_df_2["sentiment_label"].value_counts(dropna=False)
```

```python
final_df_2 = final_df_2.dropna(subset=["sentiment_label"])
final_df_2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure sentiment_label is categorical
final_df_2["sentiment_label"] = final_df_2["sentiment_label"].astype(str)

# Boxplot for Lifecycle Hours
plt.figure(figsize=(8,5))
sns.boxplot(x="sentiment_label", y="lifecycle_hours", data=final_df_2)
plt.yscale("log")  # optional if distribution is skewed
plt.xlabel("Sentiment Label")
plt.ylabel("Lifecycle Hours")
plt.title("Distribution of PR Lifecycle Hours by Sentiment")
plt.show()
```

```python

# Boxplot for Iteration Cycles
plt.figure(figsize=(8,5))
sns.boxplot(x="sentiment_label", y="iteration_cycles", data=final_df_2)
plt.xlabel("Sentiment Label")
plt.ylabel("Iteration Cycles")
plt.title("Distribution of PR Iteration Cycles by Sentiment")
plt.show()
```
